# Data Visualization with Matplotlib
Welcome to the data visualization module! In this notebook, we'll learn how to create effective visualizations using Matplotlib, Python's most popular plotting library.

## What you'll learn:
1. Creating basic plots (line, bar, scatter)
2. Customizing plot appearance
3. Creating multiple subplots
4. Saving visualizations

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set style for better-looking plots
plt.style.use('seaborn-v0_8-darkgrid')
```

## Line Plot
Let's start with a simple line plot showing monthly sales trends:

```python
# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [1200, 1500, 1400, 1700, 1600, 1800]

# Create line plot
plt.figure(figsize=(10, 6))
plt.plot(months, sales, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.show()
```

## Bar Plot
Bar plots are great for comparing categories:

```python
# Sample data
products = ['Laptop', 'Mouse', 'Keyboard', 'Monitor']
units_sold = [50, 200, 150, 75]

# Create bar plot
plt.figure(figsize=(10, 6))
plt.bar(products, units_sold)
plt.title('Units Sold by Product')
plt.xlabel('Product')
plt.ylabel('Units Sold')

# Add value labels on top of each bar
for i, v in enumerate(units_sold):
    plt.text(i, v + 5, str(v), ha='center')

plt.show()
```

## Multiple Subplots
You can create multiple plots in the same figure:

```python
# Create sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# First subplot
ax1.plot(x, y1)
ax1.set_title('Sine Wave')
ax1.grid(True)

# Second subplot
ax2.plot(x, y2)
ax2.set_title('Cosine Wave')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

## Exercise
Create a visualization that:
1. Shows both total revenue and units sold for each product
2. Uses different colors for revenue and units
3. Includes a legend
4. Saves the plot as 'sales_analysis.png'

```python
# Your code here
# Example solution:
# Read the sales data
sales_df = pd.read_csv('../2_pandas_fundamentals/sales_analysis.csv')

# Create figure
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot total revenue
color1 = 'tab:blue'
ax1.set_xlabel('Product')
ax1.set_ylabel('Total Revenue ($)', color=color1)
revenue_bars = ax1.bar(sales_df['Product'], sales_df['Total Revenue'], color=color1, alpha=0.7)
ax1.tick_params(axis='y', labelcolor=color1)

# Create second y-axis for units sold
ax2 = ax1.twinx()
color2 = 'tab:orange'
ax2.set_ylabel('Total Units Sold', color=color2)
units_line = ax2.plot(sales_df['Product'], sales_df['Total Units'], color=color2, marker='o', linewidth=2, label='Units Sold')
ax2.tick_params(axis='y', labelcolor=color2)

# Add title and adjust layout
plt.title('Product Performance Analysis')
fig.tight_layout()

# Add legend
lines1 = [plt.Rectangle((0,0),1,1, fc=color1, alpha=0.7)]
lines2 = units_line
labels1 = ['Total Revenue']
labels2 = ['Units Sold']
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

# Save the plot
plt.savefig('sales_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```
